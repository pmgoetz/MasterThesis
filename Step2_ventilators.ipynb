```python
from LastElement import get_last_element
```

```python
#define the outcomes
Model.outcomes = [ScalarOutcome('Coverage ventilator', variable_name='Total normalized coverage ventilator',
                                          kind=ScalarOutcome.MAXIMIZE, function = get_last_element)]
```

```python
#defining which convergence metrics you want to save
convergence_metrics =[EpsilonProgress()]
```

```python
#optimising the decision lever for the 1 outcomes, over 10000 runs
with SequentialEvaluator(Model) as evaluator:
    results, convergence = evaluator.optimize(nfe=4000, searchover='levers', convergence=convergence_metrics,
                                      epsilons=[0.05,] * len(Model.outcomes) , Scenario=ref_scenario)
```
